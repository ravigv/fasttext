```python
import flashtext
from flashtext import KeywordProcessor

```

```python
def file_path(filename, path):
    path = pathlib.Path(path)
    fullpath =  Path().joinpath(path, filename)
    return fullpath
    




filename = 'names.first.female.txt'
path = "F:\\NLP\\indain_names\\"


deffile_path(filename, path)
```

```python
def remove_names(txt, filename, path):
    
    file = file_path(filename, path)
    names = file.read_text().split()
    keyword_processor = KeywordProcessor()
    for name in names:
        keyword_processor.add_keyword(name, " " )
        

    new_sentence = keyword_processor.replace_keywords(txt)
#     new_sentence = ''.join(new_sentence)
    return new_sentence



filename = 'names.first.male.txt'
path = "F:\\NLP\\indain_names\\"
txt = 'Rama Raju Farhan eat'


remove_names(txt, filename, path)
```
